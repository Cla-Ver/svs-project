```python
import os
import pandas as pd
```

```python
folderName = "train"
class_id = "/m/01jfm_"

files = os.listdir(f"dataset/{folderName}/images")
bbox = pd.read_csv(f"{folderName}_csv/{folderName}-annotations-bbox.csv", header=0, usecols=["ImageID", "XMin", "YMin", "XMax", "YMax", "LabelName"])
bbox["ImageID"] = bbox["ImageID"].str.strip()
```

```python
test_img = files[0].title().split(".")[0].lower().strip()
print(test_img)
test_bbox = bbox[(bbox["ImageID"] == test_img) & (bbox["LabelName"] == class_id)]
print(test_bbox)
```

```python
for image in files:
    image_name = image.title().split(".")[0].lower().strip()
    bbox_image = bbox[(bbox["ImageID"] == image_name) & (bbox["LabelName"] == class_id)]
    
    with open(f"dataset/{folderName}/annotations/{image_name}.txt", "w") as f:
        for _, box in bbox_image.iterrows():
            #print(box)
            x_center = (box["XMin"] + box["XMax"]) / 2
            y_center = (box["YMin"] + box["YMax"]) / 2
            width = box["XMax"] - box["XMin"]
            height = box["YMax"] - box["YMin"]
            f.write(f"{x_center} {y_center} {width} {height}\n")
```
